# The notebook manipulate the dataset to produce a new one suitable for the training purpose  

```python
import pandas as pd


pd.set_option('display.max_columns', None)
```

```python
df=pd.read_csv("out/2_elaborated_dataset_with_embeddings.csv")
df
```

```python
#filter the dataframe
df=df[["name","embeddings","goal"]]
df
```

```python
df["goal"].iloc[0],type(df["goal"].iloc[0])
```

```python
# convert string of strings into list of strings
import ast
df["goal_list"] = df["goal"].apply(lambda x: ast.literal_eval(x) ) 
```

```python
df["goal_list"].iloc[0],type(df["goal_list"].iloc[0])
```

## Transform the dataset to be suitable for a multilabel classification problem
similar to one-hot encoding but with the possibility to have multiple 1 setted (multilabel)

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Initialize MultiLabelBinarizer
mlb = MultiLabelBinarizer()

# Transform the labels into a one-hot encoded format
encoded_labels = mlb.fit_transform(df['goal_list'])

# Convert the result into a DataFrame with class names as columns
encoded_df = pd.DataFrame(encoded_labels, columns=mlb.classes_)

# Concatenate the original text with the one-hot encoded labels
df_encoded = pd.concat([df[['name','embeddings']], encoded_df], axis=1)

print(df_encoded)
```

```python
# each row can have "mutiple one-hot" values for the target labels (e.g. see raw 33) 
df_encoded.head(50)
```

```python
# column "name" is not more necessary 
df_encoded.drop("name", axis=1,inplace=True)
```

```python
df_encoded
```

```python
df_encoded.to_csv("out/3_elaborated_dataset_for_multilabel_training.csv", index=False)
```
